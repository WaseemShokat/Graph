```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sugar_level=[81,80,82,91,78,77,73,92,99,89,88,87]
plt.hist(sugar_level,bins=[70,75,80,85,90,95,100], rwidth=.95)
```

```python
men_sugar_level=[81,80,82,91,78,77,73,92,99,89,88,87]
women_sugar_level=[71,70,72,71,68,87,93,72,79,99,98,77]
plt.xlabel('Sugar Level')
plt.ylabel('Total Number')
plt.title('Sugar Level Analysis')
plt.hist([men_sugar_level,women_sugar_level],bins=[70,75,80,85,90,95,100], rwidth=.95,color=['green','red'],
         label=['men','women'])
plt.legend()
plt.grid()
```

```python
men_sugar_level=[81,80,82,91,78,77,73,92,99,89,88,87]
women_sugar_level=[71,70,72,71,68,87,93,72,79,99,98,77]
plt.ylabel('Sugar Level')
plt.xlabel('Total Number')
plt.title('Sugar Level Analysis')
plt.hist([men_sugar_level,women_sugar_level],bins=[70,75,80,85,90,95,100], rwidth=.95,color=['green','red'],
         label=['men','women'],orientation='horizontal')
plt.legend()
plt.grid()
```
