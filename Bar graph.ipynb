```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
comapny=["Google","Amz","Microsoft","FB"]
revenue=[230,249,320,210]
plt.bar(comapny,revenue)
```

```python
xpos=np.arange(len(comapny))
xpos
```

```python
revenue=[230,249,320,210]
plt.bar(xpos,revenue)
```

```python
profit=[50,23,100,19]
```

```python

plt.title("US Tech Stocks")
plt.ylabel("Billion")
plt.xticks(xpos,comapny)
plt.bar(xpos-.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+.2,profit,width=0.4,label="profit")
plt.legend()
```

```python
plt.title("US Tech Stocks")
plt.yticks(xpos,comapny)
plt.barh(xpos-.2,revenue,label="Revenue")
plt.barh(xpos+.2,profit,label="profit")
plt.legend()
plt.savefig("US Tech.png")
```
