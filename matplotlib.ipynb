```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
x=[2,4,5,6,3,4,1]
y=[45,34,56,65,21,44,53]
plt.plot(x,y,linewidth=6,color='green',linestyle='dashed',marker='o',markersize=12)
```

```python
Days=[1,2,3,4,5,6,7]
Max_Temperature=[23,31,32,29,28,34,35]
Min_Temperature=[13,21,22,19,18,24,25]
Avg_Temperature=[18,28,26,24,24,29,29]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(Days,Max_Temperature,label="Max")
plt.plot(Days,Min_Temperature,label="Min")
plt.plot(Days,Avg_Temperature,label="Avg")

plt.legend(loc="best",shadow=True)
plt.grid()
```
